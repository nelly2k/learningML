# Maps
- Maps is going to run a function against all rows in a DataFrame (can look like lambda)
- doesn't transform DataFrame, just returns values

```python
hours_per_week_mean = df['hours-per-week'].mean()
df['hours-per-week'].map(lambda p: p-hours_per_week_mean)
```

```python
# apply actually transform DataFrame
def remain_hours(row):
    row.points = row['hours-per-week'] - hours_per_week_mean
    return row

df.apply(remain_hours, axis='columns')
```
